```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Helpef function to import a fle and return a dataframe with a type assertion
def import_file(file_name):
    assert(isinstance(file_name, str))
    return pd.read_csv(file_name)

locs = ['N.L.', 'P.E.I.', 'N.S.', 'N.B. ','Que.', 'Ont.', 'Nun.', 'Man.', 'Sask.', 'Alta.', 'B.C.', 'Y.T.', 'N.W.T.']
```

We need to select desired queries and choose what capitas to analyze. Since the locations and indices match, this is an easy process.

```python
expenditure_by_type = import_file("data/canada_data/expenditure_by_type.csv").dropna()

#['Year ','Hospitals', 'Other Institutions', 'Physicians',
#        'Other Professionals ', 'Drugs', 'Capital', 'Public Health',
#          'Administration', 'Other Health Spending']
#

# Queries to select and corresponding graphs to analyze
queries = ['Year','Hospitals', 'Drugs']
ind_private_percapita = ["D.2.1.3","D.2.2.3","D.2.3.3",
                         "D.2.4.3","D.2.5.3","D.2.6.3",
                         "D.2.7.3","D.2.8.3","D.2.9.3",
                         "D.2.10.3","D.2.11.3","D.2.12.3",
                         "D.2.13.3"]
ind_private = ["D.2.1.2","D.2.2.2","D.2.3.2",
                         "D.2.4.2","D.2.5.2","D.2.6.2",
                         "D.2.7.2","D.2.8.2","D.2.9.2",
                         "D.2.10.2","D.2.11.2","D.2.12.2",
                         "D.2.13.2"]
ind_public = ["F.3.1.2","F.3.2.2","F.3.3.2",
                         "F.3.4.2","F.3.5.2","F.3.6.2",
                         "F.3.7.2","F.3.8.2","F.3.9.2",
                         "F.3.10.2","F.3.11.2","F.3.12.2",
                         "F.3.13.2"]
ind_public_percap = ["F.3.1.3","F.3.2.3","F.3.3.3",
                         "F.3.4.3","F.3.5.3","F.3.6.3",
                         "F.3.7.3","F.3.8.3","F.3.9.3",
                         "F.3.10.3","F.3.11.3","F.3.12.3",
                         "F.3.13.3"]
```

These next four cells are of the same type, essentially run through categories of data and generate corresponding plots for a select group of features.

```python
for index in ind_public:
    df_bc = expenditure_by_type.loc[expenditure_by_type["Category"].str.contains(index)]
    df1 = df_bc[queries]
    ax = df1.astype(float).plot(x='Year')
    ax.set_xlabel("Year"),
    ax.set_ylabel("% use of funds towards")
    ax.set_title("% distribution of provincial government health spending in " + locs[int(index.split(".")[2])-1])
```

```python
for index in ind_public_percap:
    df_bc = expenditure_by_type.loc[expenditure_by_type["Category"].str.contains(index)]
    df1 = df_bc[queries]
    ax = df1.astype(float).plot(x='Year')
    ax.set_xlabel("Year"),
    ax.set_ylabel("Per capita health expenditure (C$)")
    ax.set_title("Provincial government spending in " + locs[int(index.split(".")[2])-1])
```

```python
for index in ind_private:
    df_bc = expenditure_by_type.loc[expenditure_by_type["Category"].str.contains(index)]
    df1 = df_bc[queries]
    df1 = df1.astype(float).dropna()
    ax = df1.plot(x='Year')
    ax.set_xlabel("Year"),
    ax.set_ylabel("% use of funds towards")
    ax.set_title("% distribution of Private-sector health spending in " + locs[int(index.split(".")[2])-1])
```

```python
for index in ind_private_percapita:
    df_bc = expenditure_by_type.loc[expenditure_by_type["Category"].str.contains(index)]
    df1 = df_bc[queries]
    df1 = df1.astype(float).dropna()
    ax = df1.plot(x='Year')
    ax.set_xlabel("Year"),
    ax.set_ylabel("Per capita health expenditure (C$)")
    ax.set_title("Private-sector spending in " + locs[int(index.split(".")[2])-1])
```
